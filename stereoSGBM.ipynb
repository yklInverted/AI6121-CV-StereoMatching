```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
left_image = cv2.imread('./images/triclopsi2l.jpg', cv2.IMREAD_GRAYSCALE)
right_image = cv2.imread('./images/triclopsi2r.jpg', cv2.IMREAD_GRAYSCALE)
left_image = np.uint8(left_image)
right_image = np.uint8(right_image)

wsize=9
max_disp = 16
sigma = 1.5
lmbda = 8000.0

left_matcher = cv2.StereoSGBM_create(0, max_disp, wsize);
right_matcher = cv2.ximgproc.createRightMatcher(left_matcher);
left_disp = left_matcher.compute(left_image, right_image);
right_disp = right_matcher.compute(right_image,left_image);

# Now create DisparityWLSFilter
wls_filter = cv2.ximgproc.createDisparityWLSFilter(left_matcher);
wls_filter.setLambda(lmbda);
wls_filter.setSigmaColor(sigma);
filtered_disp = wls_filter.filter(left_disp, left_image, disparity_map_right=right_disp);
plt.imshow(filtered_disp, cmap='hot')
```
